```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel('titanic_dataset.xlsx')

# Display the first few rows of the dataset
print(df.head())

# Summary statistics
print(df.describe())

# Information about the dataset
print(df.info())
```

```python
# Check for missing values
print(df.isnull().sum())

# Visualizing missing values
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

Here, We gather the passengerid, age, fare, sex, embarked and survived using "Heat Map".
We can easily understand the data in form of heatmap effectively.

```python
# Distribution of numerical features
df.hist(bins=30, figsize=(20,15), color='steelblue', edgecolor='black')
plt.suptitle('Histograms of Numerical Features')
plt.show()
```

In the above code we created histograms seperately for pasengerid, age, sex, fare, embarked and survived data.

```python
# Countplot of categorical features
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Survived', palette='viridis')
plt.title('Survival Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Sex', palette='viridis')
plt.title('Sex Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Embarked', palette='viridis')
plt.title('Embarked Count')
plt.show()
```

In the above code we taken countplots for "Survival count, sex count and embarked count". From this we get to know the count of eache varaible from the dataset.

```python
# Pairplot of numerical features colored by survival
sns.pairplot(df.dropna(), hue='Survived', diag_kind='hist')
plt.suptitle('Pairplot of Numerical Features by Survival', y=1.02)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

The correlation heatmap helps identify relationships between numerical features. In the Titanic dataset:

Higher fares correlate with higher survival rates, likely due to better cabin conditions.
Age has a minor negative correlation with survival, indicating younger passengers had a slightly better chance of surviving.
The relationship between Age and Fare is weak, indicating no significant age-based fare differences.

```python
# Scatter plot of Age vs Fare colored by survival
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', palette='viridis')
plt.title('Age vs Fare by Survival')
plt.show()
```

```python
# Boxplot of Age by Sex
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Sex', y='Age', palette='viridis')
plt.title('Boxplot of Age by Sex')
plt.show()
```

```python
# Violin plot of Age by Embarked
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='Embarked', y='Age', palette='viridis')
plt.title('Violin Plot of Age by Embarked')
plt.show()
```

We can analyze the distribution of ages for passengers who embarked from different locations.
The violin plot provides insights into the age distributions and highlights the differences in the age profiles of passengers based on their embarkation point.
It helps to understand the demographic differences among passengers embarking from different ports, which might influence survival rates and other factors.
